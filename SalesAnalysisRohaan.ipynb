```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
plt.style.use(['dark_background'])
```

Merging 12 months of Sales data into a single file.

```python
df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')
all_months_data = pd.DataFrame()
files = [file for file in os.listdir('./Sales_Data')]

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])
```

```python
all_months_data.to_csv("all_data.csv", index = False)
```

Read in Updated Dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

Clean the Data

Drop rows with NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how = 'all')
all_data.head()
```

```python
#Find 'Or' and delete it
all_data = all_data[all_data['Order Date'].str[:2] != 'Or']
all_data.head()
```

```python
#Convert the datatype of columns
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

Augment data with additional columns

Add month column

```python
#Take first 3 characters from Order Date
all_data['Month'] = all_data['Order Date'].str[:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

Add Sales column

```python
all_data["Sales"] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

Add a city column

```python
#Lets split on the commas and get the first index
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1] #Dont want the area code
all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x) + ' ' + get_state(x))
all_data.head()
```

What was the best month for sales? How much was earned in that month?

```python
results = all_data.groupby('Month').sum()
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month')
plt.show()
```

Which city had the highest sales

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.ylabel('Sales in USD')
plt.xlabel('City')
plt.show()
```

What time should we display the advetisements to increase the likelihood of the customer buying the product?

```python
#Convert order date to datetime object
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count()) #Count no of rows for each hour in group by
plt.xticks(hours, rotation = 'vertical', size=8)
plt.grid()
plt.ylabel('Total Orders')
plt.xlabel('Hour')
plt.show()
```

Which products are often sold together?

```python
#If 2 orders have same order id, they were bought together
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
#Joining the products sold together into a single row
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
#Count which row occured most often
from itertools import combinations
from collections import Counter
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key,value)
```

Which was the most sold product and why?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
products = [product for product, df in product_group]
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation = 'vertical', size=8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.show()
#AAA batteries were sold the most. Cheap, more common in households
```

```python
#Lets show the prices of the products in the above bar chart
prices = all_data.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, quantity_ordered, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price', color='b')
ax1.set_xticklabels(products, rotation ='vertical', size=8)
plt.show()
#We can see that since the price is low, 
#the quantity ordered in high for a few products, thus confirming our observation
```
